```python
import cv2
from PIL import Image
```

```python
max_area = 0
a = b = c = d = 0
m = 30
```

```python
%cd "D:\Official\Masters Studies\Thesis\Dataset\Segmentation\Analysis\source"
f1 = open("pred.txt","r")
f2 = open("img.txt","r")
pred = f1.read().split("\n")
img = f2.read().split("\n")
for pd,im in zip(pred,img):
    image = cv2.imread(pd)
    g_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    n_img = g_img.copy()
    n_img[g_img == 38] = 0
    contours, _ = cv2.findContours(n_img, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)
    for c in contours:
        area = cv2.contourArea(c)
        if area > 1000:
            if area > max_area:
                biggest_cnt = c
                max_area = area
                rect = cv2.boundingRect(biggest_cnt)
                x,y,w,h = rect
    if x-m >= 0 : a = x - m
    else: a = x
    if y-m >= 0 : b = y - m
    else: b = y
    c = x+w+m
    d = y+h+m
    #if x+w+m <= g_img.shape[1] : c = x+w+m
    #else: c = g_img.shape[1]
    #if y+h+m <= g_img.shape[0] : d = y+h+m
    #else: d = g_img.shape[0]
    box = (a, b, c, d)
    o_img = Image.open(im)
    c_img = o_img.crop(box)
    %cd "D:\Official\Masters Studies\Thesis\Dataset\Segmentation\Analysis\cropped"
    c_img.save(im)
    %cd "D:\Official\Masters Studies\Thesis\Dataset\Segmentation\Analysis\source"
    
```
