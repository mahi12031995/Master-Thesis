```python
import cv2
import os
%cd "D:\Official\Masters Studies\Thesis\Dataset\Bolt Dataset\selected_set\Bolt_1852\red cable workout\try\aug source"
mydir = r'D:\Official\Masters Studies\Thesis\Dataset\Bolt Dataset\selected_set\Bolt_1852\red cable workout\try\aug result'
my_bgs = open("bgs.txt", "r")
my_gts = open("gts.txt", "r")
my_imgs = open("imgs.txt", "r")
bgs = my_bgs.read().split("\n")
gts = my_gts.read().split("\n")
imgs = my_imgs.read().split("\n")
i = 0
for bg,gt,img in zip(bgs,gts,imgs):
    i += 1
    print(i)
    c_bg = cv2.imread(bg)
    c_gt = cv2.imread(gt)
    c_img = cv2.imread(img)
    n_bg = c_bg.copy()
    n_img = c_img.copy()
    n_img[c_gt == 0] = 0
    n_bg[c_gt != 0] = 0
    masked = n_img + n_bg
    cv2.imwrite(os.path.join(mydir,img), masked)
```

```python
import numpy
import cv2
%cd "D:\Official\Masters Studies\Thesis\Dataset\Segmentation\Falzenwelb\analysis 2\iou"
my_file = open("mahi.txt", "r")
data = my_file.read().split("\n")
siou = 0
miou = 0
i = 0
for d in data:
    %cd "D:\Official\Masters Studies\Thesis\Dataset\Segmentation\Falzenwelb\analysis 2\binary gt"
    result1 = cv2.imread(d)
    %cd "D:\Official\Masters Studies\Thesis\Dataset\Segmentation\Falzenwelb\analysis 2\binary marker"
    result2 = cv2.imread(d)
    intersection = numpy.logical_and(result1, result2)
    union = numpy.logical_or(result1, result2)
    iou_score = numpy.sum(intersection) / numpy.sum(union)
    siou = siou + iou_score
miou = siou/152
print(miou)
```
